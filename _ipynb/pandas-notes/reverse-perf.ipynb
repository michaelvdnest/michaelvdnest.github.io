```python
import pandas as pd
import numpy as np
import perfplot
```

```python
def _setup(n):
    
    df = pd.DataFrame({
        "A": np.random.rand(n),
        "B": np.random.rand(n),
        "C": np.random.rand(n),
    })

    return df
```

```python
perfplot.show(
    setup=_setup,
    kernels=[
        lambda df: df.sort_index(ascending=False),
        lambda df: df.iloc[::-1],
        lambda df: df.sort_index(ascending=False).reset_index(drop=True),
        lambda df: df.iloc[::-1].reset_index(drop=True),
        
    ],
    labels=["sort", "iloc", "sort reindex", "iloc reindex"],
    n_range=[2 ** k for k in range(18)],
    xlabel="len(df)",
    equality_check=None,
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# hide
df = pd.read_csv('./tidy-data/billboard.csv')
df
```

```python
# hide
cols = df.columns.tolist()
id_vars = [i for i in cols if not i.endswith('.week')]
id_vars
```

```python
# hide
# Reshape
df = df.melt(id_vars=id_vars, var_name='week', value_name='rank')
df = df.dropna()

# Formatting
df["week"] = df['week'].str.extract('(\d+)', expand=False).astype(int)
df["rank"] = df["rank"].astype(int)

df = df.sort_values(ascending=True, by=["year","artist.inverted","track","week","rank"])

df
```

```python
# hide 
df
```

```python
# hide 
df['date'] = pd.to_datetime(df['date.entered']) + pd.to_timedelta(df['week'], unit='w') - pd.DateOffset(weeks=1)
df
```

```python
# hide
df.columns = ['year', 'artist', 'track', 'time', 'genre', 'entered', 'peaked', 'week', 'rank', 'date']
df = df[["year", "artist", "track", "time", "genre", "week", "rank", "date"]]
billboard = df
```

```python
billboard
```

```python
songs_cols = ["year", "artist", "track", "time", "genre"]
songs = billboard[songs_cols].drop_duplicates()
songs = songs.reset_index(drop=True)
songs["song_id"] = songs.index
songs
```

```python
# merge song id to billboard records
ranks = pd.merge(billboard, songs, on=["year","artist", "track", "time", "genre"])
ranks
```

```python
ranks = ranks[["song_id", "date","rank"]]
ranks
```

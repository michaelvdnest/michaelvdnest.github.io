```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame(
    {
        "foo": ["one", "one", "one", "two", "two", "two"],
        "bar": ["A", "B", "C", "A", "B", "C"],
        "baz": [1, 2, 3, 4, 5, 6],
        "qux": [6, 5, 4, 3, 2, 1],
        'quux': ['a', 'b', 'b', 'c', 'c', 'c']
    }
)
```

```python
# hide 
df
```

```python
# hide
df.groupby('foo').groups
```

```python
# hide
grouped = df.groupby('foo')
for name,group in grouped:
    print(name)
    print(group)
```

```python
# hide
grouped.get_group('one')
```

```python
grouped.ngroup()
```

```python
sc = lambda x: (x - x.mean()) / x.std()*10
grouped.transform(sc)
```

```python
grouped.transform(sc)
```

```python
ngroup = pd.Series(grouped.ngroup(), name= 'foo_ngroup')
cumcount = pd.Series(grouped.cumcount(), name= 'foo_cumcount')
rank = pd.Series(grouped['baz'].rank(), name= 'foo_rank')
sc = grouped.transform(sc).rename(columns=lambda x: x + '_sc')

pd.concat([df, ngroup, cumcount, rank, sc], axis=1)
```
